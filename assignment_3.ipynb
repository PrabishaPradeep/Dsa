# **3.   Who scores the most on average for math, reading and writing based on gender**
      

```python
df.groupby('gender')[['math score','reading score','writing score']].mean()
```

```python
df.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title('Math, Reading and Writing Score Averages based on Gender')
plt.ylabel('average')
plt.show()
```

females scores most on average in both reading and writing whereas male students scores most on average in maths.

# **4.   Who scores the most on average for math, reading and writing based on Test preparation course**
      

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
```

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title('Math, Reading and Writing Score Averages based on Test Preparation Course')
plt.ylabel('average')
plt.show()
```

It seems that students that completed the test preparation course score higher than those that did not.

# **6. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.**

Top 25% of students are basically the students who scored third quartile (75th percentage) of score or more.



```python
maths_threshold =df["math score"].quantile(0.75)
maths_threshold
```
