An Institution wishes to find out their student’s ability in mathematics, reading and writing skills. The Institution wants to do an exploratory study to check the following information.

let us start by importing necessary libraries ,then read the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/StudentsPerformance.csv")
df
```

1.   Find out how many males and females participated in the test.




```python
df['gender'].value_counts()
```

2.   What do you think about the students' parental level of education?




```python
parents=df["parental level of education"].value_counts()
parents
```

```python
df['parental level of education'].value_counts(normalize = True)
df['parental level of education'].value_counts(dropna = False).plot.bar()
plt.title('analysis of Parental Education')
plt.xlabel('educational qualification')
plt.ylabel('count')
plt.show()
```

It can be seen  that parents studied in some college is highest and parents with master's degree is too low.

3.   Who scores the most on average for math, reading and writing based on
      

              1.   Gender
              2.   Test preparation course

```python
df.groupby('gender')['math score','reading score','writing score'].mean()
```

females scores most on average in both reading and writing whereas male students scores most on average in maths.

```python
df.groupby('test preparation course')['math score','reading score','writing score'].mean()
```

It seems that students that completed the test preparation course score higher than those that did not.

4. What do you think about the scoring variation for math, reading and writing based on
*   Gender
*   Test preparation course



```python
df.groupby('gender')['math score','reading score','writing score'].var()
```

females shows higher scoring variation for math, reading and writing as compared to males.

```python
df.groupby('test preparation course')['math score','reading score','writing score'].var()
```

students who have completed the test preparation course shows smaller variation in math, reading and writing as compared to those who have not.

5. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.

```python
maths_threshold=df["math score"].quantile(0.75)
top_25_percent_maths_students=df[df['math score']>= maths_threshold]
print(top_25_percent_maths_students)
```

these are the students selected for bonus points
